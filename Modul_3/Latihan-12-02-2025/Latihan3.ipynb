```python
# Inisialisasi variabel
harga_tiket = 150000
jumlah_penumpang = 5
total_harga = harga_tiket * jumlah_penumpang

# Cek diskon
if jumlah_penumpang > 4:
    persentase_diskon = 0.1
    total_harga_diskon = total_harga * (1 - persentase_diskon)
    print("Anda mendapat diskon 10%!")
    print(f"Total Harga Setelah Diskon: Rp {int(total_harga_diskon)}")
else:
    print("Tidak ada diskon.")
    print(f"Total Harga: Rp {int(total_harga)}")
```
