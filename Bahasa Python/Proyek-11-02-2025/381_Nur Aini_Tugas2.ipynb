```python
Angka_pertama = int(input("Masukkan Angka pertama: "))
Angka_kedua = int(input("Masukkan Angka kedua: "))

# Change 'angka1' and 'angka2' to 'Angka_pertama' and 'Angka_kedua'
jumlah = Angka_pertama + Angka_kedua
kurang = Angka_pertama - Angka_kedua
kali = Angka_pertama * Angka_kedua
bagi = Angka_pertama / Angka_kedua

print("Hasil penjumlahan:", jumlah)
print("Hasil pengurangan:", kurang)
print("Hasil perkalian:", kali)
print("Hasil pembagian:", bagi)
```
