```python
stock_buku = 5
jumlah_peminjaman = int(input("Masukkan jumlah buku yang ingin dipinjam: "))

if jumlah_peminjaman <= stock_buku:
    print("Peminjaman berhasil, buku tersedia.")
    stock_buku = stock_buku - jumlah_peminjaman
else: # Aligned with the 'if' statement
    print("Peminjaman gagal, stok tidak mencukupi.")
    print("Stok tersedia hanya:", stock_buku)
```
